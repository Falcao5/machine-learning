```python
from sklearn.compose import make_column_transformer
```

```python
# applico
column_trans = make_column_transformer((OneHotEncoder(), ["Club","POSITION"]), remainder="passthrough")
```

```python
array = column_trans.fit_transform(df1_2015)
```

```python
array.shape
```

```python
#distinguo due dataset, uno con le feature e uno con la classe ossia il fattore di classificazione, tutto tramite subsetting
df, cl = (array[:,0:47], df1_2015["class"])
```

```python
cl.shape
```

```python
df.shape
```

```python
df
```

```python
#creo i dataset
x_train, x_test, y_train, y_test = train_test_split(df, cl, test_size=0.3, random_state = 12345)
```

```python
#ulteriore passo di preprocessing
y_train = np_utils.to_categorical(y_train,2)
y_test = np_utils.to_categorical(y_test,2)
```

```python
#creo il modello neurale
```

```python
model = Sequential() #inizializzo il modello
```

```python
#aggiungo i layers e i neuroni
model.add(layers.Dense(45, input_dim = x_train.shape[1], activation="relu"))
#procedo aggiungendo un altro strato
model.add(layers.Dense(22, activation="relu")) #la funzione di attivazione è la rectified linear unit
model.add(layers.Dense(11, activation="relu"))
model.add(layers.Dense(5, activation="relu"))
#strato id uscita
model.add(layers.Dense(2, activation="softmax")) #2 neuroni siccome le possibili classifcazioni sono 2, la funzione softmax meglio si adatta ai probl. di classificazior multipla
```

```python
model.summary()
```

```python
#compilo il modello con funzione di perdita, ottimizzatore e metrica
model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
#applico sugli oggetti di test
score = model.evaluate(x_test, y_test)
```
